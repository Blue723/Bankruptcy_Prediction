For this notebook I will asses banks liabilities.

```python
import pandas as pd

import re

import math

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.info()
```

```python
liability_col = df.filter(regex=".*Liability.*")
liability_col.columns
```

```python
re.findall('.*owing.*', str(df.columns))
```

```python
plot_cols = liability_col.columns
ncols = 4
plot_rows=math.ceil(len(plot_cols) / ncols)



fig, ax = plt.subplots(plot_rows, ncols, figsize=(10,15))

for i in range(len(plot_cols)):
    sns.kdeplot(liability_col[plot_cols[i]], ax = ax[i // ncols, i % ncols])
    if i % ncols != 0:
        ax[i // ncols, i % ncols].set_ylabel(" ")
plt.tight_layout()
plt.show()
```
